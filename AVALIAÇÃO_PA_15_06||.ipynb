QUESTÃO 02 – Soma dos Números (2 pontos)
---------------------------------------------------------

Solicite ao usuário um número inteiro positivo.

Utilizando um laço de repetição, calcule e exiba a soma de
todos os números de 1 até o número informado.

Exemplo:

Digite um número: 5

Resultado: 15

Pois:

1 + 2 + 3 + 4 + 5 = 15

```python
while True:
    numero = int(input("Digite um número:  "))
    if numero > 0:
        break
    print("Por favor, digite um número maior que zero.")

soma = 0
passos = []

for i in range(1, numero + 1):
    soma += i
    passos.append(str(i))

explicacao = " + ".join(passos)

print("")
print(f"Resultado: {soma}")
print("")
print(f"Pois: {explicacao} = {soma}")

```

QUESTÃO 04 – Fatorial (4 pontos)
---------------------------------------------------------

Solicite ao usuário um número inteiro positivo e calcule
seu fatorial utilizando um laço de repetição.

Exemplo:

Digite um número: 5

Fatorial = 120

Pois:

5! = 5 × 4 × 3 × 2 × 1

Observação:
O programa deve validar se o número informado é positivo.


```python
numero = int(input("Digite um número:  "))

multiplica = 1
passos = []

for i in range(1, numero + 1):
    multiplica *= i
    passos.append(str(i))

explicacao = " x ".join(passos)

print("")
print(f"||Fatorial: {multiplica}||")
print("")
print(f"Pois: |{explicacao} = {multiplica}|")

if multiplica < 0:
  print("")
  print("O resultado é negativo.")

elif multiplica > 0:
  print("")
  print("O resultado é positivo.")
```

QUESTÃO 05 – Caixa Eletrônico (5 pontos)
---------------------------------------------------------

Um caixa eletrônico possui notas de:


R$ 100

R$ 50

R$ 20

R$ 10

R$ 5

R$ 2

R$ 1
__________________________________________________________
Solicite ao usuário o valor do saque e informe a menor
quantidade possível de notas para realizar o pagamento.
__________________________________________________________
Exemplo:

Digite o valor do saque: 287

_________________________________
Saída esperada:
---------------------------------
2 nota(s) de R$100

1 nota(s) de R$50

1 nota(s) de R$20

1 nota(s) de R$10

1 nota(s) de R$5

1 nota(s) de R$2


```python
# Solicita o valor do saque ao usuário
a = int(input("Digite o valor do saque: "))
print("")


# Define a lista de notas disponíveis em ordem decrescente
notas = [100, 50, 20, 10, 5, 2, 1]

print("-" * 33)
print("Notas necessárias:")
print("-" * 33)
print("")

# Percorre cada uma das notas disponíveis
for nota in notas:

    # Calcula a quantidade de notas necessárias da cédula atual
    quantidade_notas = a // nota

    # Atualiza o valor restante para as próximas notas
    a %= nota

    # Exibe o resultado apenas se a quantidade de notas for maior que zero
    if quantidade_notas > 0:
        print(f"-- {quantidade_notas} nota(s) de R${nota}")
```
