#Escrever um programa que o usuário digita seu nome e idade e o programa retorna se ele é adulto, adolescente ou criança.

```python
#Entrada

nome = str (input("Digite o seu nome:  "))
idade = int (input("Digite sua idade:  "))

# Processamento e Saída

if idade < 12:
  print(f"Seja bem vindo(a), {nome}. Você é uma criança.")

elif idade < 18:
  print(f"Seja bem vindo(a), {nome}. Você é adolescente.")

elif idade > 60:
  print(f"Seja bem vindo(a), {nome}. Você é idoso.")

else:
  print(f"Seja bem vindo(a), {nome}. Você é adulto.")
```

#Escrever um programa que classifica os triângulos.

* Verificar se forma um triângulo válido.
* Tipo: Equilátero, Isósceles e Escaleno.



```python
#Entrada

print("|------------------------------------------------------------------------|")
print("| Olá, seja bem vindo(a)! Hoje, você irá digitar lados de um triângulo e | ")
print("| eu verificarei se o triângulo é: Equilátero, Isósceles ou Escaleno!    |")
print("|------------------------------------------------------------------------|")
print("")
print("")
lado1 = int (input("Por favor, digite o 1º lado:  "))
lado2 = int (input("Agora, digite o 2º lado:  "))
lado3 = int (input("Digite o 3º lado:  "))
certo = lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1
print("")

# Processamento


if certo == False:
  print("|-------------------------------|")
  print("| O seu triângulo não é válido. |")
  print("|-------------------------------|")

elif certo == True and lado1 == lado2 == lado3:
  print("|--------------------------------------------------------------|")
  print("| Os lados são iguais, logo, este é um triângulo {equilatero}. |")
  print("|--------------------------------------------------------------|")

elif certo == True and lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
  print("|--------------------------------------------------------------|")
  print("| Os lados são diferentes, logo, este é um triângulo escaleno. |")
  print("|--------------------------------------------------------------|")

else:
  print("|-------------------------------------------------------------|")
  print("| Dois lados são iguais, logo, este é um triângulo isósceles. |")
  print("|-------------------------------------------------------------|")
```
