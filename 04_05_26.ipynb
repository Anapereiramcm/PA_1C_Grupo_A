##Escrever um programa que o usuário digita um número e o programa imprime de 0 até o número.

```python
número = int(input("Digite um número:  "))
print("")

for i in range (número + 1 ):
  print(i)

```

##Escrever um programa que o usuário digita um número e o programa imprime em ordem decrescente até o zero.

```python
número = int (input("Digite um número:  "))

print("")
for i in range (número, -1, -1):
  print(i)
```

##Peça um número e mostre todos os números pares até o número.

```python
coisa = int(input("Digite um número:  "))
print("")
print(f"Os números pares de 0 até {coisa} são:" )

for i in range (coisa + 1 ):
  if i % 2 == 0:
    print (i)
```

## Peça um número e calcule a soma de 1 até o número.

```python
número = int(input("Digite um número:  "))
print("")
print(f"a soma de 1 até  {número} é: ")
print("")

for i in range (número + 1):
  print (f"1 + {i} = {1 + i}")
```

## Peça uma palavra e conte as letras, ex.: Python - A Palavra tem 6 letras

```python
palavra = str(input("Digite uma palavra:  "))
contar = len(f"{palavra}")
print("")

print (f"A palavra ''{palavra}'' tem {contar} letras.")
```

## Peça uma palavra e conte as vogais.

```python
palavra = str (input("Digite uma palavra:  "))
contar = 0
vogais = "a, e, i, o, u, A, E, I, O, U, ã, á, à, ó, õ, ô, â"
print("")

for char in palavra:
  if char in vogais:
    contar += 1

print(f"A palavra ''{palavra}'' tem {contar} vogais.")
```

## Peça ao usuário para digitar 5 números e informe o maior.

```python
num1 = int(input("Digite um número:  "))
num2 = int(input("Digite um número:  "))
num3 = int(input("Digite um número:  "))
num4 = int(input("Digite um número:  "))
num5 = int(input("Digite um número:  "))

numeros = (num1, num2, num3, num4, num5)

# Encontra o maior número manualmente
maior = numeros[0]
for num in numeros:
    if num > maior:
        maior = num


print("")
print (f"O maior número digitado foi {maior}.")
```
