Escrever um algoritmo e um programa que recebe 04 notas do aluno e calcula a média, depois mostrá-la.

# Entrada

ler nota1, nota2, nota3 e nota4

# Processamento

Calcular média das notas

#Saída

Mostrar a média

```python
# Entrada

matemática = 10
português = 0
inglês = 3
história = 10

# Processamento

média = (matemática + português + inglês + história)/4


# Saída

print(f"A média do aluno é igual a: {média}")
```
