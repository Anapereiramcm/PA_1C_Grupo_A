#Contagem de 0 até 5

```python
# Com for:

for i in range (6):
  print(i)
```

```python
#Com while

contador = 0

while contador <= 5:
  print(contador)
  contador += 1 # contador += 1 é a mesma coisa que contador = contador + 1

```

#De 5 até 0

```python
contador = 5

while contador >= 0:
  print(contador)
  contador -= 1 # contador -= 1 é a mesma coisa que contador = contador - 1
```
