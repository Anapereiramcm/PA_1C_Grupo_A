_Interações e laços de repetição_

* Loop

```python
#repetir um bloco de código um número específico de vezes,
#gerando uma sequência numérica.

#Simplificar:

# "i" é uma variável local.

for i in range (5):
  print ("Olá mundo !! ")
```

```python
for i in range (5):
  print (i)
```

```python
# Não quero que apareça o zero

for i in range (1,6):
  print(i)
```

```python
#Para imprimir ao contrário (6 até 0)

for i in range (6, -1, -1 ):
  print (i)
```

```python
# Outro jeito de imprimir ao contrário (6 até 0)

for i in range (7):
  print(6 - i)
```

```python
#Quero imprimir ao contrário sem o zero.

for i in range (6):
  print (6 - i)
```

# Escrever um prorama que imprime os números pares de 0 a 10.



```python
for i in range (6):
  print (i * 2)
```

```python
#CORREÇÃO

for i in range (11):
  if i % 2 == 0:
    print(i)
```

#Imprimir números ímpares

```python
for i in range (1, 11, 2):
  print(i)
```

#Escrever um progama que o usuário digita um número de 1 a 10 e imprime a tabuada

```python
a = int(input("Digite um número de 1 a 10:  "))

for i in range (1, 11):
    print(f"| {a} x {i} = {a * i} |")
```
