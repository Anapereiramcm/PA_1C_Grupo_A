Escrever um programa e verificar se o usuário é maior de idade.

```python
# Entrada

nome = str (input("Digite seu nome:  "))
idade = int (input("Digite sua idade:  "))
print("")

# Processamento

if idade > 18:
  print(f"{nome}, você teve o acesso permitido ao site.")

else:
  print(f"{nome}, você teve seu acesso negado ao site, devido à sua idade de {idade} anos..")
```

#Escrever um programa que verifica se o número é par ou ímpar.

```python
# Entrada

a = int (input("Digite um número:  "))
print(f"")

# Procesamento

if a % 2 == 0:
  print(f"O número {a} é um número par.") #Saída

else:
  print(f"O número {a} é um número ímpar.") #Saída
```

#Escreva um programa que:
Peça o valor da compra

Se o valor for maior que 100, aplique um desconto de 10%.

Se não, mostre o valor normal.

```python
# Entrada

valor = float (input("Digite o valor da compra:  R$"))
desconto = valor * 0.90

# Processamento

if valor > 100:
  print(f"O valor final da sua compra será R${desconto}. ")

else:
  print(f"O valor final da sua compra será R${valor}.")
```

# Escrever um programa que o aluno digita 04 notas dos bimestres e o programa calcula a média e mostra, na saída, se o aluno foi aprovado, recuperação ou reprovado.

```python
# Entrada

nota1 = float (input("Digite a nota do 1º bimestre:  "))
nota2 = float (input("Digite a nota do 2º bimestre:  "))
nota3 = float (input("Digite a nota do 3º bimestre:  "))
nota4 = float (input("Digite a nota do 4º bimestre:  "))
print("")

# Processamento

média = (nota1 + nota2 + nota3 + nota4) / 4

if média >= 7:
  print(f"O aluno está aprovado. Sua média foi de {média}.")

elif  média >=5:
    print(f"O aluno está de recuperação pela média1 de {média}.")

else:
  print(f"O aluno foi reprovado pela média de {média}.")
```
