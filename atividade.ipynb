# Questão 1 – Classificação de idade

 Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:

	 •	Até 17 anos: Menor de idade

	 •	De 18 a 59 anos: Adulto

	 •	60 anos ou mais: Idoso


```python
# Entrada

nome= str (input("Digite o seu nome:  "))
idade = int (input("Digite a sua idade:  "))
print("")

# Processamento e Saída

if idade <= 17:
    print(f"Seja bem vindo(a), {nome}, você é menor de idade.")
    
elif idade <= 59: 
    print(f"Seja bem vindo(a), {nome}, você está na idade adulta.")
    
else:
    print(f"Seja vem vindo(a), {nome}, você é idoso(a).")
```

# Questão 2 – Cálculo de desconto

 Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:

  •	Até R$100: sem desconto

  •	De R$101 a R$500: 10% de desconto

  •	Acima de R$500: 20% de desconto
  
 Ao final, mostre o valor final da compra.


```python
# Entrada

valor = float (input("Digite o valor de sua compra:  "))
dez = valor * 0.9
vinte = valor * 0.8

# Processamento e Saída

if valor <= 100:
    print(f"O valor de R${valor} não terá descontos.")
    print(f"Valor final: R${valor}")

elif valor <= 500:
    print(f"O valor de R${valor} terá um desconto de 10%.")
    print(f"Valor final: R${dez}")

else: 
    print(f"O valor de {valor} terá um desconto de 20%.")
    print(f"Valor final: R${vinte}")
```

# Questão 3 – Verificação de triângulo
 Crie um programa que leia três valores e verifique se eles podem formar um triângulo.

 Para isso, a soma de dois lados deve ser maior que o terceiro.

 Mostre uma mensagem informando se é ou não um triângulo válido.
 

```python
# Entrada 

um = float (input("Digite o valor de um lado de um triângulo:  "))
dois = float (input("Digite o valor de um segundo lado de um triângulo:  "))
tres = float (input("Digite o valor de um terceiro lado de um triângulo:  "))

# Processamento e saída

if um + dois > tres and um + tres > dois and tres + dois > um:
    print("O seu triângulo é válido! ")
    print("Comemore! :)")

else: 
    print("O seu triângulo não é válido.")
    print("Chore.")
```

# Questão 4 – Sistema de energia (jogo)

 Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:

	 •	0: Morto

	 •	1 a 30: Crítico

	 •	31 a 70: Normal

	 •	71 a 100: Saudável

```python
# Entrada 

nome = str (input("Digite o nome do seu personagem: "))
energia = int (input("Digite a energia do personagem (0 à 100):  "))

# Processamento e Saída

if energia == 0:
    print(f"{nome} está morto(a)!")

elif energia <= 30:
    print(f"{nome} está com a energia em estado crítico!")

elif energia <= 70:
    print(f"{nome} está com a energia normal.")

elif energia <= 100:
    print(f"{nome} está saudável!")

else: 
    print(f"O valor de {energia} é inválido.")
```

# Questão 5 – Média escolar

 Crie um programa que leia a média final de um aluno e classifique:

 •	Média maior ou igual a 7: Aprovado

 •	Média entre 5 e 6.9: Recuperação

 •	Média menor que 5: Reprovado

```python
# Entrada

nome = str (input("Digite o nome do(a) aluno(a):  "))

mat = float (input("Digite a nota de matemática:  "))
fisica = float (input("Digite a nota de física:  "))
bio = float (input("Digite a nota de biologia:  "))
pt = float (input("Digite a nota de português: "))
ef = float (input("Digite a nota de Educação Física:  "))

media = (mat + fisica + bio + pt + ef) / 5

# Processamento e Saída

if media >= 7:
    print(f"A média de {nome} foi de {media}")
    print("")
    print(f"{nome} está aprovado(a)!")

elif media > 5:
    print(f"A média de {nome} foi de {media}")
    print("")
    print(f"{nome} está de recuperação. ")

else:
    print(f"A média de {nome} foi de {media}")
    print("")
    print(f"{nome} está reprovado(a).")

    
```

# Questão 6 – Cálculo de salário com bônus

 Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras:
 
	 •	Pontuação menor que 5: sem bônus

	 •	Pontuação de 5 a 7: bônus de 5% do salário

	 •	Pontuação de 8 a 9: bônus de 10% do salário

	 •	Pontuação igual a 10: bônus de 15% do salário


```python
# Entrada

sal = float (input("Digite o seu salário mensal:  "))
dot = float (input("Digite a sua pontuação de desempenho recebida:  "))

cinco_sete = sal * 1.05
oito_nove = sal * 1.1
dez = sal * 1.15

# Processamento

if dot < 5:
    print(f"O seu salário é de R${sal}, e sua pontuação é de {dot}.")
    print(f"O seu salário final será de R${sal}!")

elif dot <= 7:
    print(f"O seu salário é de R${sal}, e sua pontuação é de {dot}.")
    print("Parabéns! Você receberá um bônus de 5%!")
    print(f"O seu salário final será de R${cinco_sete}!")

elif dot <= 9:
    print(f"O seu salário é de R${sal}, e sua pontuação é de {dot}.")
    print("Parabéns! Você receberá um bônus de 10%!")
    print(f"O seu salário final será de R${oito_nove}!")

elif dot == 10:
    print(f"O seu salário é de R${sal}, e sua pontuação é de {dot}.")
    print("Parabéns! Você receberá um bônus de 15%!")
    print(f"O seu salário final será de R${dez}!")

else: 
    print(f"A pontuação de desempenho {dot} é inválida. Lembre-se de que ela vai até 10.")
```
